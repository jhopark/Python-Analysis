# folium

This dataset describes the listing activity and metrics in NYC, NY for 2019.
<small>data source: <a href="https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data">Airbnb data source</a></small>

```python
import folium
import pandas as pd
```

```python
# read New York airbnb data
df = pd.read_csv("AB_NYC_2019.csv")

# print df data first 5 items
df.head()
```

```python
df.shape
```

```python
df["neighbourhood_group"].unique()
```

```python
# Number of data should be less than 1500
select_data = df[df["neighbourhood_group"] == "Manhattan"].head(1500)

# print select_data shape
select_data.shape
```

```python
map = folium.Map(
    location = [df["latitude"].mean(), df["longitude"].mean()],
    zoom_start = 11)

for n in select_data.index:
    
    loc_name = select_data.loc[n, "name"]
    
    icon_color = 'red'
    
    folium.Circle(
        location = [select_data.loc[n, "latitude"], select_data.loc[n, "longitude"]],
        popup = loc_name,
        radius = 10,
        color = icon_color,
        fill = True,
        fill_color = icon_color
    ).add_to(map)

map
```
